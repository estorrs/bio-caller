```python
def get_headerless_vcf_df(f_obj, stop_on='#CHROM'):
    line = ''
    header = ''
    while True:
        line = f_obj.readline()
        header += line
        if line[:6] == stop_on:
            break
            
    df = pd.read_csv(f_obj, sep='\t', header=None)
    df.columns = line[1:].replace('""', '').replace('\n', '').split('\t')

    return df, header
```

```python
def get_chrom_pos_lookup_table(df):
    # make lookup table
    position_lookup = {}
    for i, row in df.iterrows():
        if row['CHROM'] in position_lookup:
            d = position_lookup[row['CHROM']]
            d[row['POS']] = i
        else:
            position_lookup[row['CHROM']] = {
                row['POS']: i
            }
    
    return position_lookup
```

```python
def get_base_counts(bamrc_row):
    counts = {}
    for i in range(extra_cols):
        if pd.isnull(bamrc_row[i]):
            break

        pieces = bamrc_row[i].split(':', 2)
        counts[pieces[0]] = int(pieces[1])
    
    return counts

def call_genotype(alt_allele, bamrc_row, N_thresh=.9, min_read_depth=8):
    # check read depth
    if bamrc_row['DEPTH'] < min_read_depth:
        return '.|.'
    
    base_counts = get_base_counts(bamrc_row)
    
    # check n
    if base_counts['N'] / bamrc_row['DEPTH'] > N_thresh:
        return '.|.'
    
    ref_count = base_counts.get(bamrc_row['REF'])
    alt_count = base_counts.get(alt_allele)
    
    if ref_count is None or alt_count is None:
        return '.|.'
    
    
    # check hom ref
    if ref_count >= 8 and alt_count < 4:
        return '0|0'
    # check het
    if ref_count >= 4 and alt_count >=4:
        return '0|1'
    # check hom alt
    if alt_count >= 8 and ref_count < 4:
        return '1|1'
    
    return '.|.'
```

```python
def call_for_bamrc_fp(bamrc_fp, vcf_fp):
    vcf_df, header = get_headerless_vcf_df(open(vcf_fp))
    
    extra_cols = 50 # pad columns because bam-readcount gives variable column lengths
    columns = ['CHROM', 'POS', 'REF', 'DEPTH'] + list(range(extra_cols))
    readcount_df = pd.read_csv(bamrc_fp, sep='\t', header=None, names=columns)
    # check for chr in chromosome
    if 'chr' in readcount_df['CHROM'][0]:
        readcount_df['CHROM'] = [c.replace('chr', '') for c in readcount_df['CHROM']]
    
    position_lookup = get_chrom_pos_lookup_table(readcount_df)
    
    calls = []
    for i, vcf_row in vcf_df.iterrows():
    #     if i % 1000 == 0:
    #         print i

        readcount_row = None
        d = position_lookup.get(str(vcf_row['CHROM']))
        if d:
            row_index = d.get(vcf_row['POS'])
            if row_index:
                readcount_row = readcount_df.iloc[row_index]

        if readcount_row is not None:
            calls.append(call_genotype(vcf_row['ALT'], readcount_row))
        else:
            calls.append('.|.')

    return (bamrc_fp, calls)

def proccess(arg):
    print(arg)
    return call_for_bamrc_fp(arg[0], arg[1])
```

```python
def rebuild_vcf(input_vcf_fp, sample_to_genotype):
    vcf_df, header = get_headerless_vcf_df(open(input_vcf_fp))
    
    # chop off samples
    vcf_df = vcf_df[['CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO']]
    
    # add format col
    vcf_df['FORMAT'] = ['GT'] * vcf_df.shape[0]
    
    # add samples
    for sample_id, calls in sample_to_genotype.items():
        vcf_df[sample_id] = calls
        
    # reform header
    new_header = ''
    for line in header.split('\n'):
        if line[:6] != '#CHROM':
            new_header += line + '\n'
        else:
            break
    vcf_df.rename(columns={'CHROM':'#CHROM'}, inplace=True)
        
    # combine header and df
    df_str = vcf_df.to_csv(None, index=None, sep='\t')
    output_str = new_header + df_str
        
#     return output_str
    return output_str
```

```python
WKDIR = '/app/Projects/Users/estorrs/data/dan/data/'

vcf_fp = WKDIR + '1000-genomes/all.grch38.coding.10000.vcf'

readcounts_folder = WKDIR + 'readcount_outputs/ccrc/'

sample_readcount_fps = os.listdir(readcounts_folder)
sample_names = [s.split('.')[0] for s in sample_readcount_fps]
sample_readcount_fps = [readcounts_folder + s for s in sample_readcount_fps]
samples_fp_to_name = {s:n for s, n in zip(sample_readcount_fps, sample_names)}
```

```python
results = []
threads = 5
with Pool(threads) as p:
    results = p.map(proccess, zip(sample_readcount_fps, [vcf_fp] * len(sample_readcount_fps)))
```

```python
results = {samples_fp_to_name[x[0]]:x[1] for x in results}
```

```python
df_str = rebuild_vcf(vcf_fp, results)

f = open(WKDIR + 'ccrc.10000.vcf', 'w')
f.write(df_str)
f.close()
```

```python
def coverage(calls):
    c = 0
    for call in calls:
        if call != './.':
            c += 1
    
    if len(calls) == 0:
        return 0.0
    
    return float(c) / len(calls)
```
